An exercise to plot the various components of the CPI according to three cateogiries of 'high (red)', 'medium (yellow)' and 'low (green)'

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap, BoundaryNorm
df = pd.read_excel('dispersion chart.xlsx', sheet_name= "Broad Components", header=0)
```

```python
# List of columns to plot
components = ['Food and beverages', 'Pan, tobacco and intoxicants',
              'Clothing and footwear', 'Housing',
              'Fuel and light', 'Miscellaneous', 'CPI']

# Corresponding category columns
categories = ['Food Band', 'Pan Band', 'Clothing and Footwear Band',
              'Housing Band', 'Fuel and Light Band', 'Misc Band', 'CPI Band']
```

```python
#----bubble chart broad snapshot----
colors = {'High':'red', 'Moderate':'orange', 'Low':'green'}

# Define thresholds for classification
threshold_low = 1
threshold_high = 3

# Create 'Band' columns based on inflation rates for each component
# Exclude the first row (index 0) which contains 'Weight' in 'Dates' and other metadata
df_filtered = df.iloc[1:].copy()

# Convert 'Dates' column to datetime, coercing errors, and then drop rows with NaT
df_filtered['Dates'] = pd.to_datetime(df_filtered['Dates'], errors='coerce')
df_filtered.dropna(subset=['Dates'], inplace=True)

for comp, cat in zip(components, categories):
    # Ensure the component column exists before trying to classify it
    if comp in df_filtered.columns:
        df_filtered[cat] = pd.cut(df_filtered[comp],
                         bins=[-float('inf'), threshold_low, threshold_high, float('inf')],
                         labels=['Low', 'Moderate', 'High'],
                         right=False) # 'right=False' makes bins like [low, high)
    else:
        print(f"Warning: Component column '{comp}' not found in DataFrame. Skipping band creation for it.")

plt.figure(figsize=(12, 8))
# Adding the threshold lines at 1% and 3%
plt.axhline(y=1, color='purple', linestyle='--', linewidth=1.5, label='1% Inflation Threshold')
plt.axhline(y=3, color='blue', linestyle='--', linewidth=1.5, label='3% Inflation Threshold')
for comp, cat in zip(components, categories):
    for level in ['High', 'Moderate', 'Low']:
        subset = df_filtered[df_filtered[cat] == level]
        plt.scatter(subset['Dates'], subset[comp],
                    color=colors[level], label=f'{comp} - {level}')

# To avoid duplicate legends and include the new threshold lines
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Inflation Rate (%)')
plt.title('Dispersion of Inflation Categories Across Components')
plt.tight_layout()
plt.show()
```

```python
#---line charts---

#--- different colours for different components----
components = ['Food and beverages', 'Pan, tobacco and intoxicants',
              'Clothing and footwear', 'Housing',
              'Fuel and light', 'Miscellaneous','CPI']

thresholds = [1, 3]  # Low <1%, Moderate 1-3%, High >3%
colors = ['blue', 'orange', 'green', 'purple', 'brown', 'cyan', 'black']  # line colors

plt.figure(figsize=(14,7))

# Create a cleaned DataFrame for plotting, similar to how df_filtered was created previously
df_plot = df.iloc[1:].copy()
df_plot['Dates'] = pd.to_datetime(df_plot['Dates'], errors='coerce')
df_plot.dropna(subset=['Dates'], inplace=True)

# Plot each component
for comp, color in zip(components, colors):
    plt.plot(df_plot['Dates'], df_plot[comp], marker='o', label=comp, color=color)

# Threshold lines
plt.axhline(y=thresholds[0], color='green', linestyle='--', label='Low/Moderate Threshold (1%)')
plt.axhline(y=thresholds[1], color='red', linestyle='--', label='Moderate/High Threshold (3%)')

# Shaded regions for the inflation band 4  2%
plt.fill_between(df_plot['Dates'], 0, thresholds[0], color='green', alpha=0.05)
plt.fill_between(df_plot['Dates'], thresholds[0], thresholds[1], color='yellow', alpha=0.05)
plt.fill_between(df_plot['Dates'], thresholds[1], 6, color='red', alpha=0.05)

# Reduce y-axis scale to focus on relevant range
plt.ylim(-6,10)  # Set min and max values of y-axis

# Aesthetics
plt.xticks(rotation=45)
plt.ylabel('Inflation Rate (%)')
plt.title('Inflation Rates of All Components with Thresholds (Scaled)')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Define distinct marker styles for each component
# This will help differentiate components directly on the plot!
marker_styles = ['o', 's', '^', 'D', 'X', 'P', '*'] # Circle, Square, Up-triangle, Diamond, X, Plus, Star
component_markers = dict(zip(components, marker_styles))

# Define colors for categories
colors = {'High':'red', 'Moderate':'orange', 'Low':'green'}

plt.figure(figsize=(12, 8))

# Adding the threshold lines for 1% and 3% inflation
plt.axhline(y=1, color='purple', linestyle='--', linewidth=1.5, label='1% Inflation Threshold')
plt.axhline(y=3, color='blue', linestyle='--', linewidth=1.5, label='3% Inflation Threshold')


# Using the color (for level) and marker style (for component)
for comp_index, (comp, cat) in enumerate(zip(components, categories)):
    current_marker = component_markers[comp] # Get the specific marker for this component
    for level in ['High', 'Moderate', 'Low']:
        subset = df_filtered[df_filtered[cat] == level]
        # Plotting with both color for 'level' and marker for 'comp'
        plt.scatter(subset['Dates'], subset[comp],
                    color=colors[level],
                    marker=current_marker,
                    label=f'{comp} - {level}')



# collecting all unique labels and their handles for the legend.
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Inflation Rate (%)')
plt.title('Dispersion of Inflation Categories Across Components')
plt.tight_layout()
plt.show()
```

A heatmap of the inflation rates (last year - current year) for different components

```python
plt.figure(figsize=(12,6))
sns.heatmap(df[components].T, cmap="RdYlGn_r", annot=True)
plt.title('Heatmap of Inflation Rates by Component')
plt.show()
```

```python

# Create a continuous colormap that goes green  yellow  red
colors = ["green", "yellow", "red"]
cmap = LinearSegmentedColormap.from_list("inflation_cmap", colors)

# Define your thresholds for normalization
bounds = [0, 1, 3, df[components].max().max()]  # min=0, 1, 3, max value in data
norm = BoundaryNorm(bounds, cmap.N, clip=True)

plt.figure(figsize=(12,6))
sns.heatmap(df[components].T,
            cmap=cmap,
            norm=norm,
            annot=True,
            linewidths=0.5,
            linecolor='gray')

plt.title('Heatmap of Inflation Rates with Threshold-based Gradient')
plt.show()
```

```python
---#disrtibution according to the weights----
weights = {
    'Food and beverages': 45.87,
    'Pan, tobacco and intoxicants': 2.38,
    'Clothing and footwear': 6.53,
    'Housing': 6.84,
    'Fuel and light': 10.07,
    'Miscellaneous': 28.31
}

components = list(weights.keys())
category_colors = {'Low':'green', 'Moderate':'yellow', 'High':'red'}

# Use the predefined categories list to get the correct category column names
categories = ['Food Band', 'Pan Band', 'Clothing and Footwear Band',
              'Housing Band', 'Fuel and Light Band', 'Misc Band', 'GPI Band']


plt.figure(figsize=(14,8))

# Loop over components
for i, comp in enumerate(components):
    # Find the corresponding category column name from the categories list
    try:
        cat_col = categories[components.index(comp)]
    except ValueError:
        # Handle cases where a component might not have a corresponding category column
        print(f"Warning: No category column found for component '{comp}'. Skipping.")
        continue

    for level in ['Low','Moderate','High']:
        subset = df_filtered[df_filtered[cat_col] == level]
        plt.scatter(subset['Dates'],
                    [i]*len(subset),                     # y-position for each component
                    s=subset[comp]*weights[comp],        # bubble size scaled by weight
                    c=category_colors[level],            # color by category
                    alpha=0.6, edgecolors='k',
                    label=f"{comp} - {level}")

# Improve aesthetics
plt.yticks(range(len(components)), components)
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.title('Bubble Chart of Inflation Rates by Component Weighted by Contribution')

# Custom legend for categories
legend_elements = [Patch(facecolor='green', label='Low (<1%)'),
                   Patch(facecolor='yellow', label='Moderate (1-3%)'),
                   Patch(facecolor='red', label='High (>3%)')]
plt.legend(handles=legend_elements, title='Inflation Band', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
